```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import PorterStemmer
nltk.download('punkt')
```

```python
data=pd.read_csv('/content/HateSpeechDetection.csv')
data.head()
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def tokenize_sentence(sentence):
    doc = nlp(sentence)
    tokens = [token.text for token in doc]
    return tokens

data['Tokenized_Comment'] = data['Comment'].apply(tokenize_sentence)

# Display the DataFrame with tokenized comments
print(data)
```

```python
def tokenize_sentence(sentence):
    doc = nlp(sentence)
    lemmas = [token.lemma_ for token in doc]
    return lemmas

data['Lemmatized_Comment'] = data['Comment'].apply(tokenize_sentence)

# Display the DataFrame with tokenized comments
print(data)
```
